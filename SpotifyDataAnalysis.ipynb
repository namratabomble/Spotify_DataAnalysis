```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv("C:/Users/Namrata Bomble/Downloads/archive (3)/tracks.csv")
df_tracks.head()
```

```python
#find 5 least popular songs
least_popular = df_tracks.sort_values('popularity', ascending="True").head(5)
least_popular
```

```python
#find the song which have popularity greater than 95
#how to run a query on dataset

df_tracks.query('popularity>95')
```

```python
#how to make some column as index of dataset
#inplace is for making changes in original dataset
df_tracks.set_index("release_date",inplace= True)
df_tracks.index=pd.to_datetime(df_tracks.index)
```

```python
#to check the particular value through column name and row index
df_tracks[["artists"]].iloc[10]
```

```python
#to perform some change in row values
#at last new column got created names as "duration_in_sec" 
df_tracks["duration_in_sec"]= df_tracks["duration_ms"].apply(lambda x:round(x/1000))
df_tracks.head()
```

#how to delete(drop) any column
df_tracks.drop("mode",inplace=True,axis=1)
df_tracks.head()

```python
#taking 5 percent of total dataset
sample_data = df_tracks.sample(int(0.004*len(df_tracks)))
print(len(sample_data))
```

```python
plt.figure(figsize=(10,8))
sns.regplot(data=sample_data, y= "loudness", x="energy", color= "c").set(title="loudness vs energy Correlation")
```

```python
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates= pd.to_datetime(df_tracks.dates)
year = df_tracks.dates.dt.year
```

```python
#plot Histogram

sns.displot(year,discrete=True,aspect=2,height=5,kind="hist")
```
